```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May","5-10")
print(df.head())
```

```python
# Gender Distribution
```

```python

plt.figure(figsize=(5,5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analyzed that, the number of female 
#is more than the number of male
```

```python
#analyzing the data , in  which the parent education impacts the
#average marks of students or not
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean","ReadingScore": "mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parentsEducation and Scores")
plt.show()
```

```python
#from the above chart, we have concluded that the education
#of parents impacts the avergae score of students,higher the 
#education gets the good average marks as compare to lower education
#degree level
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python

plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parentMaritalStatus and Scores")
plt.show()
```

```python
#from the above chart we have analyzed that the mariatal status 
#have no impact on student score or 
#negligible impact on student score
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
print(mlist)
plt.pie(mlist, labels = l,autopct= '%1.2f%%')
plt.title('Distribution of Ethnic Groups')
plt.show()
```
